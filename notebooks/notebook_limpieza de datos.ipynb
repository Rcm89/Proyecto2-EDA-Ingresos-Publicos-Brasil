Queremos averiguar el porcentaje de nulos en cada columna del dataFrame

```python
df_porcentaje_nulos = mostrar_porcentaje_nulos(dfs_concatenados)
df_porcentaje_nulos
```

El análisis de valores nulos muestra que las columnas con más valores nulos son:

    - VALOR PREVISTO ACTUALIZADO: 5% de valores nulos.
    - VALOR REALIZADO: 3.8% de valores nulos.
    - ORIGEN DE INGRESO: 3.7% de valores nulos.
    - TIPO DE INGRESO: 3.1% de valores nulos.

Para tratar estos valores nulos vamos a, por un lado imputar las columnas numéricas con la mediana, y eliminar las filas con valores nulos en las columnas categóricas, dado que una pequeña pérdida de datos (<5%) no debería afectar el análisis global.

```python
# Imputamos con la mediana en las columnas "VALOR PREVISTO ACTUALIZADO" y "VALOR REALIZADO"
dfs_concatenados['VALOR PREVISTO ACTUALIZADO'].fillna(dfs_concatenados['VALOR PREVISTO ACTUALIZADO'].median(), inplace=True)
dfs_concatenados['VALOR REALIZADO'].fillna(dfs_concatenados['VALOR REALIZADO'].median(), inplace=True)

# Eliminamos las filas con valores nulos en columnas categóricas
columnas_categoricas = ['ORIGEN DE INGRESO', 'TIPO DE INGRESO', 'CATEGORÍA ECONÓMICA']
dfs_concatenados.dropna(subset=columnas_categoricas, inplace=True)

# Comprobamos el porcentaje de nulos de nuevo:
df_porcentaje_nulos = mostrar_porcentaje_nulos(dfs_concatenados)
df_porcentaje_nulos

```
